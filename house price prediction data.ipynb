```python
import pandas as pd
import numpy as np
import sklearn
```

```python
data1=pd.read_csv("C:\\Users\\Shivdas\\Downloads\\datasets_46927_85203_data.csv")
```

```python
data1.head()
```

```python
# independent variable
x=data1.iloc[:,2:13]
x.head()
```

```python
# dependent variable
y=data1.iloc[:,1:2]
y.head()
```

### Spliting data in train and test , fitting data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,test_size=0.20,random_state=100)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
knn=KNeighborsRegressor(n_neighbors=8)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(y_pred[1])
y_test.head(1)
```

### RMSE

```python
for i in range(1,15):
    knn=KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    MSE=np.sqrt(mean_squared_error(y_test,y_pred))
    RMSE=sqrt(MSE)
    print("RMSE when k is"+str(i)+"=",RMSE)
```

### Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
for i in range (1,15):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train , y_train)
    y_pred = knn.predict(x_test)
    MAE =  mean_absolute_error(y_test,y_pred)
    print("MAE when k is"+str(i)+"=",MAE)
```
